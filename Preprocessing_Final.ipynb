```python
import numpy as np
import pandas as pd
import gc

train = pd.read_csv("train_ver2.csv", low_memory=False,skipinitialspace=True,
                   dtype= {  
                               "sexo":str,
                               "ind_nuevo":str,
                               "ult_fec_cli_1t":str,
                               "indext":str
                          }
                    )

train=train.drop(columns='nomprov')
train.isna().sum()
```

```python
""" Consistent dtype (int) for outputs train.iloc[:,24:]
    and replace missing values with 0
"""

# {colName: 'ind_nomina_ult1', missing_values: 16063, input_dtype: 'float', desired_dtype: 'int'}
train["ind_nomina_ult1"].fillna(0,inplace=True)
train["ind_nomina_ult1"] = train["ind_nomina_ult1"].astype(int) 

# {colName: 'ind_nom_pens_ult1', missing_values: 16063, input_dtype: 'float', desired_dtype: 'int'}
train["ind_nom_pens_ult1"].fillna(0,inplace=True)
train["ind_nom_pens_ult1"] = train["ind_nom_pens_ult1"].astype(int)

"""Treat conyuemp and ult_fec_cli_1t - High missing values"""
train["conyuemp"].fillna('N',inplace=True)
#train["conyuemp"].unique()

#train["ult_fec_cli_1t"].notna() = '1'
train.loc[train["ult_fec_cli_1t"].notna(), "ult_fec_cli_1t"] = '1'
train["ult_fec_cli_1t"].fillna('0',inplace=True)
train["ult_fec_cli_1t"] = train["ult_fec_cli_1t"].astype(int) 

train.isna().sum()
```

```python
train['renta']=np.log10(train['renta'])
train['renta'].interpolate(method='linear',inplace=True)

train.loc[((train['age']>100))]['age']=np.nan
train['age'].interpolate(method='linear',inplace=True)
train.isna().sum()
```

```python
train['indrel_1mes']=train['indrel_1mes'].replace('P',5)
train['indrel_1mes']=pd.Series(train['indrel_1mes'].astype(float), dtype="category")
train['indrel_1mes']=train['indrel_1mes'].replace('nan',np.nan)
train['indrel_1mes']=(train.indrel_1mes.cat.codes.replace(-1, np.nan)
                   .interpolate().astype(int).astype('category')
                   .cat.rename_categories(train.indrel_1mes.cat.categories)).astype(int)
train.isna().sum()
```

```python
train['tiprel_1mes']=pd.Series(train['tiprel_1mes'].astype(str), dtype="category")
train['tiprel_1mes']=train['tiprel_1mes'].replace('nan',np.nan)
train['tiprel_1mes']=(train.tiprel_1mes.cat.codes.replace(-1, np.nan)
                   .interpolate().astype(int).astype('category')
                   .cat.rename_categories(train.tiprel_1mes.cat.categories))
train.isna().sum()
```

```python
train['canal_entrada']=pd.Series(train['canal_entrada'].astype(str), dtype="category")
train['canal_entrada']=train['canal_entrada'].replace('nan',np.nan)
train['canal_entrada']=(train.canal_entrada.cat.codes.replace(-1, np.nan)
                   .interpolate().astype(int).astype('category')
                   .cat.rename_categories(train.canal_entrada.cat.categories))
train.isna().sum()
```

```python
train['segmento']=pd.Series(train['segmento'].astype(str), dtype="category")
train['segmento']=train['segmento'].replace('nan',np.nan)
train['segmento']=(train.segmento.cat.codes.replace(-1, np.nan)
                   .interpolate().astype(int).astype('category')
                   .cat.rename_categories(train.segmento.cat.categories))
train.isna().sum()
```

```python
train['cod_prov']=pd.Series(train['cod_prov'].astype(str), dtype="category")
train['cod_prov']=train['cod_prov'].replace('nan',np.nan)
train['cod_prov']=(train.cod_prov.cat.codes.replace(-1, np.nan)
                   .interpolate().astype(int).astype('category')
                   .cat.rename_categories(train.cod_prov.cat.categories))
train.isna().sum()
```

```python
train.iloc[:,:23]=train.iloc[:,:23].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
test = pd.read_csv("test_ver2.csv",skipinitialspace=True)
"""Treat conyuemp and ult_fec_cli_1t - High missing values"""
test["conyuemp"].fillna('N',inplace=True)
#test["conyuemp"].unique()

#test["ult_fec_cli_1t"].notna() = '1'
test.loc[test["ult_fec_cli_1t"].notna(), "ult_fec_cli_1t"] = '1'
test["ult_fec_cli_1t"].fillna('0',inplace=True)
test["ult_fec_cli_1t"] = test["ult_fec_cli_1t"].astype(int) 

test=test.apply(lambda x: x.fillna(x.mode()[0]))
test=test.drop(columns='nomprov')
test.isna().sum()
```

```python
train.set_index('ncodpers',inplace=True)
test_train_cols= [i for i in train.columns if '_ult1' in i]
test=test.merge(right=train.loc[(train['fecha_dato']=='2016-04-28'),
                        test_train_cols],how='left',left_on='ncodpers', right_index=True)
train.reset_index(inplace=True)
test=test.sort_index(axis=1)
test=test.drop(columns='fecha_dato')
test.set_index('ncodpers',inplace=True,drop=True)
test.columns=[i+'_0' for i in test.columns]
test.info(verbose=True)
```

```python
train.set_index('ncodpers',inplace=True)
temp=test.merge(right=train.loc[(train['fecha_dato']=='2016-05-28')],
                how='left',left_index=True, right_index=True).iloc[:,45:]
train.reset_index(inplace=True)
temp=temp.sort_index(axis=1)
temp.columns=[i+'_1' for i in temp.columns]
test=pd.concat([test,temp],axis=1)
test=test.drop(columns='fecha_dato_1')
test.info(verbose=True)
#test.set_index('ncodpers',inplace=True,drop=True)
del temp 
gc.collect()
```

```python

test.to_csv('test_final_order2.csv',index=True)
del test
gc.collect()
```

```python
""" Prepare train and test for order n
"""

months = ['2015-01-28', '2015-02-28', '2015-03-28', '2015-04-28', '2015-05-28',
         '2015-06-28', '2015-07-28', '2015-08-28', '2015-09-28', '2015-10-28',
         '2015-11-28', '2015-12-28', '2016-01-28', '2016-02-28', '2016-03-28',
         '2016-04-28', '2016-05-28']

train_X= pd.DataFrame()
train_Y= pd.DataFrame()

order=2

for i in range(len(months)-order+1):
    X=train.loc[train['fecha_dato'].isin(months[i:i+order])]
    X=X.pivot(index='ncodpers', columns='fecha_dato').dropna()
    X.sort_index(axis=1,level=1,inplace=True)
    X.columns = [(col[0]+'_'+str(months[i:i+order].index(col[1]))) 
                 for j,col in enumerate(X.columns.values)]
    X=pd.DataFrame(train['ncodpers'].unique(),columns=['ncodpers']).merge(X,
                        how='left',left_on='ncodpers',right_on='ncodpers').iloc[:,1:]
    Ycols=[k for k in X.columns if 'ult1_'+str(order-1) in k]
    print(months[i:i+order])
    if months[i:i+order][0]!=months[0]:
        train_Y=train_Y.append(X.loc[:,Ycols], ignore_index=True)
        print('Y')
    if months[i:i+order][0]!=months[-order]:
        train_X=train_X.append(X, ignore_index=True)
        print('X')
    Xind=list(X.index)
```

```python
del train
gc.collect()
```

```python
X_complete=train_X.dropna().index
Y_complete=train_Y.dropna().index
Complete=X_complete&Y_complete
```

```python
train_Y=train_Y.loc[Complete]
train_X=train_X.loc[Complete]
```

```python
Ynot1 = (train_Y == 0).all(axis=1)
train_Y=train_Y[~Ynot1]
train_X=train_X[~Ynot1]
```
